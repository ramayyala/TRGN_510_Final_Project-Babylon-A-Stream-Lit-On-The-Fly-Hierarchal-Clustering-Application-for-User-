```python
import datatable as dt
import pandas as pd
import numpy as np
import glob 
```

```python
dfs = dt.iread("../data/FPKMs/*FPKM.txt",
               columns={"gene_id", "FPKM"})
```

```python
frames = []
for filename, frame in dfs:
    #print(filename)
    #mm = re.search(r"(\w+)\.sorted\.bam", filename)
    frame.names = {"FPKM": str(filename)}
    frames.append(frame)
```

```python
frames[1]
```

```python
df1 = dt.cbind(frames, force=True)
df1
```

```python
df1=df1.to_pandas()
```
