```python
import datatable as dt
import pandas as pd
import numpy as np
import glob 
```

```python
clinical = pd.read_excel (r'../metadata/exposure_clinical_sample_familyhistory.xlsx', sheet_name='clinical')
exposure = pd.read_excel (r'../metadata/exposure_clinical_sample_familyhistory.xlsx', sheet_name='exposure')
sample = pd.read_excel (r'../metadata/exposure_clinical_sample_familyhistory.xlsx', sheet_name='gdc_sample')
```

```python
clinical_exposure=pd.merge(clinical, exposure, on=["case_id","case_submitter_id","project_id"],how="right")
clinical_exposure = clinical_exposure.rename(columns={'case_id': 'Case ID'})
clinical_exposure
```

```python
clin_expo_no_dup=clinical_exposure.drop_duplicates(subset='case_submitter_id', keep="last")
clin_expo_no_dup
```

```python
sample
```

```python
sample = sample.rename(columns={'Case ID': 'case_submitter_id'})
metadata=pd.merge(sample,clin_expo_no_dup,on=["case_submitter_id"],how="left")
metadata
```

```python
metadata["gender"]
```

```python
files = [file for file in glob.glob("../data/FPKMs/*.FPKM.txt")]
```

```python
file_list=[]
for i in files:
    #print(i)
    file_list.append(i.split("../data/FPKMs/")[1]+'.gz' )
```

```python
file_names = pd.DataFrame(file_list,columns =['File Name'])
```

```python
file_names
```

```python
sample_names = list(["Sample_" + str(i).zfill(2) for i in range(1,447)])
```

```python
sample_names
```

```python
file_names = pd.DataFrame(zip(file_list,sample_names),columns =['File Name','Sample Name'])
```

```python
file_names
```

```python
covariates=pd.merge(metadata,file_names,on=["File Name"],how="left")
```

```python
covariates=covariates.sort_values(by=['Sample Name'])
```
